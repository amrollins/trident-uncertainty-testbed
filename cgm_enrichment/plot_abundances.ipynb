```python
import matplotlib.pyplot as plt
```

```python
data = np.genfromtxt("cgm_abundances.txt")
ref = data[1,:]
for i in range(1,data.shape[0]):
    plt.semilogy((data-ref)/ref*100, ls="-")
plt.ylabel("Abundance Deviation [%]")
plt.xlabel("Atomic Number")
```

```python
for i in range(25):
    data = np.genfromtxt(f"cgm_dm_{i:03d}.txt")
    plt.semilogy(np.linspace(0, 13, data.size), data, ls="-")
plt.xlim(0,13)
plt.ylabel("log DM Mass")
plt.xlabel("Time [Gyr]")
```
